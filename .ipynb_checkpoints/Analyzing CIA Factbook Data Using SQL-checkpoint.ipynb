```python
# selecting first 5 rows from the Facts table
%%sql
SELECT * FROM facts LIMIT 5;
```

```python
# selecting min & max population along with the min & max population growth
%%sql
SELECT MIN(population), MAX(population), MIN(population_growth), MAX(population_growth) FROM facts
```

```python
# find country with minimum population
%%sql
SELECT name
FROM facts
WHERE population = (SELECT MIN(population) FROM facts)
```

```python
# find country with the max population
%%sql
SELECT name
FROM facts
WHERE population = (SELECT MAX(population) FROM facts)
```

```python
# find average values for population and area columns
%%sql
SELECT AVG(population), AVG(area)
FROM facts
```

```python
# find countries where population is greater
%%sql
SELECT name
FROM facts
WHERE population > (SELECT AVG(population) from facts) 
AND (area < (SELECT AVG(area) from facts))
```
